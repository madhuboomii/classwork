```python
movies=['vtk','kgf','vikram','master','iyan','kaithi','naanee varuven']
rating=[85,45,90,40,95,90,80]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# scatter plot

```python
plt.scatter(movies,rating)
```

```python
plt.bar(movies,rating)
```

```python
plt.plot(movies,rating ,linestyle='dotted')
```

# pie chart

```python
plt.pie(rating,labels=movies)
```


# Relationship plot

```python
sb.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day",style="day",kind="line")
```

# Categorigal box plot

```python
sb.catplot(x='day',y='total_bill',kind='box',data=tip)
```

# Regression plot

```python
sb.regplot(x='total_bill',y='tip',data=tip)
```
